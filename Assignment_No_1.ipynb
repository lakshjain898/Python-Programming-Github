# **Part 2: Basic Python Syntax**

## *I) String Operations*

```python
firstname = input("Enter First Name: ")
lastname = input("Enter Last Name: ")

firstname = firstname.title() # changing the case of the first letters just in case they are input in small
lastname = lastname.title()

result = firstname + " " + lastname
print(f"Original Name: {result}")

reversed_fname = ''.join(reversed(firstname)) # reversed() returns an iterable that traverses the string in reverse order
reversed_lname = ''.join(reversed(lastname)) # join() method joins the iterated string elements using the '' empty string
result = reversed_fname + " " + reversed_lname
print(f"Reversed Name: {result}")
```

## *II) Numeric Datatypes and Conversion Functions*

```python
import math

# Take input from the user
user_input = input("Enter a number: ")

# Convert the input to different numeric types

# 1. Integer (int)
# The int data type represents whole numbers without any decimal points.
# If the input can be converted to an integer, it will be stored as such.
try:
    int_value = int(user_input)
    print(f"Integer: {int_value}")
except ValueError:
    print("Cannot convert to an integer.")
    try:
        # Attempt to convert the input to float first
        float_value = float(user_input)

        # Round the float value to the nearest integer
        rounded_value = round(float_value)
        print(f"Rounded to nearest integer: {rounded_value}")

        # Floor the float value to the nearest lower integer
        floored_value = math.floor(float_value)
        print(f"Floored to lower integer: {floored_value}")
    except ValueError:
        print("Cannot convert to a float, so rounding and flooring are not possible.")

# 2. Float (float)
# The float data type represents real numbers with decimal points.
# This conversion allows numbers like 3.14, 2.0, or even '42' (which becomes 42.0).
try:
    float_value = float(user_input)
    print(f"Float: {float_value}")
except ValueError:
    print("Cannot convert to a float.")

# 3. Complex (complex)
# The complex data type represents complex numbers in the form of a + bj, where:
# 'a' is the real part (a float), and 'b' is the imaginary part (a float).
# If the input is a number, it can be converted to a complex number with the imaginary part being 0 by default.
try:
    complex_value = complex(user_input)
    print(f"Complex: {complex_value}")
except ValueError:
    print("Cannot convert to a complex number.")
```

## *V) The % Method and print Function*

```python
def calculateAverage(numbers):
  total_sum = sum(numbers)
  return total_sum/3

while True:
  try:
    l = [int(input(f"Enter element {i+1}: ")) for i in range(3)]
    break
  except ValueError:
    print("Invalid input. Please enter integer values only")

avg = calculateAverage(l)
print("Average of %d, %d and %d is: %.2f" % (l[0], l[1], l[2], avg))
print(f"Average of 3 numbers is: {avg:.3f}")
```

# **Part 3: Language Components**

## *I) Control Flow (if Statements and Loops)*

```python
while True:
  num = input("Enter a Integer: ")
  if num.lower() == "exit":
    break
  try:
    num = int(num)
    if num > 0:
      print(f"{num} is Positive")
    elif num == 0:
      print(f"{num} is Zero")
    elif num < 0:
      print(f"{num} is Negative")
  except ValueError:
    continue
```

## *II) Relational and Logical Operators*

```python
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Check if both numbers are even
both_even = (num1 % 2 == 0) and (num2 % 2 == 0)

# Check if both numbers are odd
both_odd = (num1 % 2 != 0) and (num2 % 2 != 0)

# Check if one number is even and the other is odd
one_even_one_odd = (num1 % 2 == 0) and (num2 % 2 != 0) or (num1 % 2 != 0) and (num2 % 2 == 0)

if both_even:
    print("Both numbers are even.")
elif both_odd:
    print("Both numbers are odd.")
elif one_even_one_odd:
    print("One number is even and the other is odd.")
else:
    print("Invalid input.")
```

## *III) For Loop and Bitwise Operators*

```python
def to_binary(n):
    # Handle special case for zero
    if n == 0:
        return "0"
    binary_str = ""
    # Loop until n is 0
    while n > 0:
        binary_str = str(n & 1) + binary_str  # Get the least significant bit
        n >>= 1  # Right shift n by 1 to process the next bit
    return binary_str

def to_octal(n):
    # Handle special case for zero
    if n == 0:
        return "0"
    octal_str = ""
    while n > 0:
        octal_str = str(n & 7) + octal_str  # Get the last 3 bits (octal digit)
        n >>= 3  # Right shift n by 3 to process the next set of 3 bits
    return octal_str

def to_hexadecimal(n):
    # Handle special case for zero
    if n == 0:
        return "0"
    hex_str = ""
    hex_digits = "0123456789ABCDEF"  # Hexadecimal digits
    while n > 0:
        hex_str = hex_digits[n & 15] + hex_str  # Get the last 4 bits (hexadecimal digit)
        n >>= 4  # Right shift n by 4 to process the next set of 4 bits
    return hex_str

# Take integer input from the user
num = int(input("Enter an integer: "))

# Convert and print the number in binary, octal, and hexadecimal
binary_representation = to_binary(num)
octal_representation = to_octal(num)
hexadecimal_representation = to_hexadecimal(num)

print(f"Binary: {binary_representation}")
print(f"Octal: {octal_representation}")
print(f"Hexadecimal: {hexadecimal_representation}")
```
